```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

def build_model():
    conv_base = VGG16(
        weights="imagenet",
        include_top=False,
        input_shape=(224, 224, 3)
    )

    # Freeze layers up to block5_conv1
    conv_base.trainable=True
    
    set_trainable = False
    for layer in conv_base.layers:
        if layer.name == "block5_conv1":
            set_trainable = True
        layer.trainable = set_trainable

    # Custom head
    x = GlobalAveragePooling2D()(conv_base.output)
    x = Dense(256, activation="relu")(x)

    plant_output = Dense(13, activation="softmax", name="plant_output")(x)
    disease_output = Dense(29, activation="softmax", name="disease_output")(x)

    model = Model(
    inputs=conv_base.input,
    outputs={
        "plant_output": plant_output,
        "disease_output": disease_output
    }
)

    return model
```
